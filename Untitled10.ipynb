```python
import sqlite3

# Connect to (or create) the SQLite database
conn = sqlite3.connect("sales_data.db")
cur = conn.cursor()

# Create the sales table
cur.execute("""
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product TEXT,
    quantity INTEGER,
    price REAL
)
""")

# Insert sample sales data
sales_data = [
    ("Apple", 10, 30),
    ("Banana", 15, 10),
    ("Orange", 12, 20),
    ("Apple", 5, 30),
    ("Banana", 8, 10),
    ("Orange", 10, 20),
    ("Grapes", 6, 40),
]

cur.executemany("INSERT INTO sales (product, quantity, price) VALUES (?, ?, ?)", sales_data)

# Save and close connection
conn.commit()
conn.close()

print("Database and data created successfully ✅")
```

```python
import pandas as pd
import sqlite3

# Connect to the database
conn = sqlite3.connect("sales_data.db")

# SQL query to get total quantity and revenue per product
query = """
SELECT 
    product, 
    SUM(quantity) AS total_quantity, 
    SUM(quantity * price) AS revenue 
FROM sales 
GROUP BY product
"""

# Execute the query and load into a DataFrame
df = pd.read_sql_query(query, conn)
conn.close()

# Print the results
print("Sales Summary Table:")
print(df)
```

```python
import matplotlib.pyplot as plt

# Bar chart: Revenue by Product
df.plot(kind='bar', x='product', y='revenue', legend=False, color='skyblue')
plt.title("Revenue by Product")
plt.ylabel("Revenue (₹)")
plt.xlabel("Product")
plt.tight_layout()
plt.savefig("sales_chart.png")  # Saves chart as PNG
plt.show()
```
